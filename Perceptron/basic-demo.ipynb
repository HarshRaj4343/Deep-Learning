```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv("/Users/harshraj/Desktop/Deep-Learning/Datasets/placement.csv")
```

```python
df.sample(5)
```

```python
sns.scatterplot(data=df, x="cgpa", y="resume_score", hue="placed")
plt.show()
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1] 
```

```python
X
```

```python
y
```

```python
p = Perceptron()
p.fit(X,y)
p.score(X,y) # returns the mean accuracy on the given test data and labels.
```

```python
p.coef_
```

```python
p.intercept_
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
